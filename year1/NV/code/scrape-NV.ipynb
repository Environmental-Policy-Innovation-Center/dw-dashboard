### Appendix C

Here we use a saved file of the PDF because the tables we need are printed vertically so they need to be flipped horizontally and then saved locally for the parsing to work as intended.

```python
nv_pdfs = camelot.read_pdf(path+"/2022_DWSRF_2nd_Amendment_Intended_Use_Plan.pdf", 
                           pages='37', flavor='stream', row_tol=10)
print(len(nv_pdfs))
```

```python
nv_1 = nv_pdfs[0].df

# drop column rows
nv_1 = nv_1.iloc[2:,].reset_index(drop=True).copy()

# assign column names
nv_1.columns = ['Grant applied to:', '*PL Class', 'PL Rank', 'Community Name/Project Sponsor', 'Equivalency', 'PWS ID',
                'Population', 'Project Description', 'Total Assistance', 'Meets Disadvantaged Definition', 'drop10', 'Base Disadvantaged Subsidy',
                'drop12', 'Base Additional Subsidy', 'drop14', 'Estimated Binding Commitment Date', 'Population less than 10,000']

# drop empty columns from parsing
nv_1 = nv_1.drop(columns=['drop10', 'drop12', 'drop14'])

# replace linebreaks with spaces
nv_1['Community Name/Project Sponsor'] = nv_1['Community Name/Project Sponsor'].str.replace("\n", " ")
nv_1['Project Description'] = nv_1['Project Description'].str.replace("\n", " ")

nv_1
```
