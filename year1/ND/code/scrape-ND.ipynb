## North Dakota

```python
nd_fund = camelot.read_pdf(path+"/ND_IUPandPPL_2023.pdf", 
                           pages='15-17', flavor='lattice')
print(len(nd_fund))
```

### General Supplemental Additional Subsidization

```python
nd_supp = nd_fund[0].df

# drop line breaks in column names, then set as columns and drop first row
nd_supp.iloc[0,] = nd_supp.iloc[0,].str.replace("\n", "")
nd_supp.columns = nd_supp.iloc[0,]
nd_supp = nd_supp.iloc[1:,].copy()

nd_supp
```

### Emerging Contaminants Additional Subsidization

```python
nd_ec = nd_fund[1].df

# drop line breaks in column names, then set as columns and drop first row
nd_ec.iloc[0,] = nd_ec.iloc[0,].str.replace("\n", "")
nd_ec.columns = nd_ec.iloc[0,]
nd_ec = nd_ec.iloc[1:,].copy()

nd_ec
```

### Lead Additional Subsidization

```python
nd_lead = pd.concat([nd_fund[2].df,
                    # drop the column row to create one list of projects
                    nd_fund[3].df.loc[1:,]]).reset_index(drop=True)

# drop line breaks in column names, then set as columns and drop first row
nd_lead.iloc[0,] = nd_lead.iloc[0,].str.replace("\n", "")
nd_lead.columns = nd_lead.iloc[0,]
nd_lead = nd_lead.iloc[1:,].copy()

nd_lead
```

### Export to CSV

```python
# write all files to csv

nd_supp.to_csv("../data/year1/csv/34-NorthDakota_GeneralSupplementalFunding.csv", index=False)

nd_ec.to_csv("../data/year1/csv/34-NorthDakota_EmergingContaminantsFunding.csv", index=False)

nd_lead.to_csv("../data/year1/csv/34-NorthDakota_LeadFunding.csv", index=False)
```
